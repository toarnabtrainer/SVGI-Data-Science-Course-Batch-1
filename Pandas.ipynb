* **It is often said that 80% of data analysis activity is spent on the data cleaning and preparation of our data. To get the grip over this activity, this section will focus on a small but important aspect of data manipulation and cleaning with Pandas.**

## **Inbuilt Data Structure in Pandas**
**There are two types of data structures are there in Pandas -** <br>

* **Series -** It is one-dimensional labeled array capable of holding any data type (integer, string, float, Python objects etc.) of data. The axis is collectively referred to as index. <br>

* **Data Frame -** It is a two-dimensional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or SQL Table or a Series of objects <br>

## **Series Data Structure**
**pandas.core.series.Series(data, index, dtype, copy)** <br>
* **data -** data takes various forms like ndarray, list, constants, dictionary etc. <br>
* **index -** it is unique and hashable for easy identification. <br>
* **dtype -** it is for data type. <br>
* **copy -** copy data, and its default value is False. It only affects for Series or one dimensional ndarray inputs. <br>

```python
# importing required modules
import numpy as np
import pandas as pd
```

```python
# creating empty series
s = pd.Series()
print(s, type(s))
```

```python
# to ignore warning messages (optional)
import warnings
warnings.filterwarnings('ignore')
```

```python
# creating series from ndarray
arr_data = np.array(['apple', 'banana', 'pineapple', 'coconut'])
s = pd.Series(data = arr_data)
print(s)
print(type(s))
print(s[0], s[3])
print(s[1:3])
```

```python
# creating series from ndarray
arr_data = np.array(['apple', 'banana', 'pineapple', 'coconut'])
my_index = ['fruit-1', 'fruit-2', 'fruit-3', 'fruit-4']
s = pd.Series(data = arr_data, index = my_index)
print(s)
print(type(s))
print(s['fruit-1'], s[0], s['fruit-3'], s[2])
s[0] = 'orange'
s['fruit-3'] = "guava"
print(s)
```

```python
arr_data = np.array([100, 200, 300, 400])
s = pd.Series(arr_data, copy = False)
s[0] = 999; arr_data[2] = 888
print(arr_data)
print(s)
```

```python
arr_data = np.array([100, 200, 300, 400])
s = pd.Series(arr_data, copy = True)
s[0] = 999; arr_data[2] = 888
print(arr_data)
print(s)
```

```python
# creating series from ndarray
arr_data = np.array(['apple', 'banana', 'pineapple', 'coconut'])
my_index = ['fruit-1', 'fruit-2', 'fruit-1', 'fruit-2']
s = pd.Series(data = arr_data, index = my_index)
print(s)
print(type(s))
print(s['fruit-1'])
print(s['fruit-2'])
```

```python
# creating series from dictionary
dict_data = {'apple':100, 'banana':230, 'kiwi':450, 'lemon':345}
print(dict_data)
s = pd.Series(dict_data)
print(s)
s = pd.Series(dict_data, index = ['banana', 'apple', 'kiwi', 'lime', 'coconut'])
print(s)
```

```python
# creating series from a scalar
s = pd.Series(500, index = [100, 104, 102, 106])
print(s)
```

```python
# create series from a list
s = pd.Series(data = [100, 300, 200, 400], index = ['red', 'green', 'blue', 'black'])
print(s)
```

## DataFrame Data Structure

### Create DataFrame

```python
user_data = [['alice',19,'F','student'],['john',26,'M','student']]
user_columns = ['name', 'age', 'gender', 'job']
user1 = pd.DataFrame(data = user_data)
user1
```

```python
user_data = [['alice',19,'F','student'],['john',26,'M','student']]
user_columns = ['name', 'age', 'gender', 'job']
user1 = pd.DataFrame(data = user_data, columns = user_columns)
user1
```

```python
user_data = dict(name=['eric','paul'], age=[22,58], gender=['M','F'], job=['student','manager'])
print(user_data)
user2 = pd.DataFrame(data = user_data)
user2
```

```python
user_data = {'name':['peter','julie'], 'age':[33,44], 'gender':['M','F'], 'job':['engineer','scientist']}
print(user_data)
user3 = pd.DataFrame(data = user_data)
user3
```
