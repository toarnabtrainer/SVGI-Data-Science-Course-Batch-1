# **Numpy: Arrays and Matrices**

```python
import numpy as np
```

## **Create arrays**

```python
# create arrays from lists
# remember arrays will have data of same data types
data1 = [11, 22, 44, 33, 66, 55]
print(data1, type(data1), len(data1), id(data1))
arr1 = np.array(data1)
print(arr1, type(arr1), len(arr1), id(arr1))
```

```python
# create arrays from lists
# remember arrays will have data of same data types
data1 = [11, True, 22, 44, False, 33, 66, 55]
print(data1, type(data1), len(data1), id(data1))
arr1 = np.array(data1)
print(arr1, type(arr1), len(arr1), id(arr1))
```

```python
# create arrays from lists
# remember arrays will have data of same data types
data1 = [11.5, True, 22, 44, False, 33, 66, 55]
print(data1, type(data1), len(data1), id(data1))
arr1 = np.array(data1)
print(arr1, type(arr1), len(arr1), id(arr1))
```

```python
# create arrays from lists
# remember arrays will have data of same data types
data1 = ["11.5", True, 22, 44.5, False, 33, 66, 55]
print(data1, type(data1), len(data1), id(data1))
arr1 = np.array(data1)
print(arr1, type(arr1), len(arr1), id(arr1))
```

```python
list1 = range(1, 10, 2)
print(list1)
arr1 = np.array(list1)
print(arr1)
```

```python
# list2 = [range(1, 10), range(11, 20)] # list of lists
list2 = [[11, 22, 33, 44], [66, 77, 88, 99]]
print(list2)
arr2 = np.array(list2) # 2-D array
print(arr2)
for item in np.nditer(arr2):  # iterating over the ndarray in a row-major fashion
    print(item, end = ", ")
```

```python
# function with variable number of arguments
def funct1(*arg):  # arg is defined forcefully as tuple type
    print(arg, len(arg), type(arg))
funct1("Amit", "Developer")
funct1("Amit", "Developer", "Pune")
funct1("Amit", "Developer", "Pune", "IBM")
```

```python
# function with variable number of arguments
def funct2(**kwarg):  # kwarg is defined forcefully as dictionary type
    print(kwarg, len(kwarg), type(kwarg))
funct2(name="Amit", desig="Developer")
funct2(name="Amit", desig="Developer", location="Pune")
funct2(name="Amit", desig="Developer", location="Pune", company="IBM")
```

```python
# function with variable number of arguments
def funct3(*arg, **kwarg):
    print(arg, len(arg), type(arg))
    print(kwarg, len(kwarg), type(kwarg), "\n")

funct3("Amit", "Developer", location="Pune", company="IBM", salary=80000)
funct3("Amit", "Developer")
funct3(location="Pune", company="IBM", salary=80000)
```
