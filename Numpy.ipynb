# **Numpy: Arrays and Matrices**

```python
import numpy as np
```

## **Create arrays**

```python
# create arrays from lists
# remember arrays will have data of same data types
data1 = [11, 22, 44, 33, 66, 55]
print(data1, type(data1), len(data1), id(data1))
arr1 = np.array(data1)
print(arr1, type(arr1), len(arr1), id(arr1))
```

```python
# create arrays from lists
# remember arrays will have data of same data types
data1 = [11, True, 22, 44, False, 33, 66, 55]
print(data1, type(data1), len(data1), id(data1))
arr1 = np.array(data1)
print(arr1, type(arr1), len(arr1), id(arr1))
```

```python
# create arrays from lists
# remember arrays will have data of same data types
data1 = [11.5, True, 22, 44, False, 33, 66, 55]
print(data1, type(data1), len(data1), id(data1))
arr1 = np.array(data1)
print(arr1, type(arr1), len(arr1), id(arr1))
```

```python
# create arrays from lists
# remember arrays will have data of same data types
data1 = ["11.5", True, 22, 44.5, False, 33, 66, 55]
print(data1, type(data1), len(data1), id(data1))
arr1 = np.array(data1)
print(arr1, type(arr1), len(arr1), id(arr1))
```

```python
list1 = range(1, 10, 2)
print(list1)
arr1 = np.array(list1)
print(arr1)
```

```python
# list2 = [range(1, 10), range(11, 20)] # list of lists
list2 = [[11, 22, 33, 44], [66, 77, 88, 99]]
print(list2)
arr2 = np.array(list2) # 2-D array
print(arr2)
for item in np.nditer(arr2):  # iterating over the ndarray in a row-major fashion
    print(item, end = ", ")
```

```python
# function with variable number of arguments
def funct1(*arg):  # arg is defined forcefully as tuple type
    print(arg, len(arg), type(arg))
funct1("Amit", "Developer")
funct1("Amit", "Developer", "Pune")
funct1("Amit", "Developer", "Pune", "IBM")
```

```python
# function with variable number of arguments
def funct2(**kwarg):  # kwarg is defined forcefully as dictionary type
    print(kwarg, len(kwarg), type(kwarg))
funct2(name="Amit", desig="Developer")
funct2(name="Amit", desig="Developer", location="Pune")
funct2(name="Amit", desig="Developer", location="Pune", company="IBM")
```

```python
# function with variable number of arguments
def funct3(*arg, **kwarg):
    print(arg, len(arg), type(arg))
    print(kwarg, len(kwarg), type(kwarg), "\n")

funct3("Amit", "Developer", location="Pune", company="IBM", salary=80000)
funct3("Amit", "Developer")
funct3(location="Pune", company="IBM", salary=80000)
```

```python
# nd-array to list
list1 = [11, 33, 22, 55, 44, 66]
print(list1, len(list1), type(list1))
arr1 = np.array(list1)
print(arr1, len(arr1), type(arr1))
list2 = arr1.tolist()
print(list2, len(list2), type(list2))
list2 = list(arr1)  # not to be done
print(list2, len(list2), type(list2))
```

```python
# nd-array to list
list1 = [[11, 33, 22, 55], [44, 66, 88, 77]] # list of lists
print(list1, len(list1), type(list1))
arr1 = np.array(list1)
print(arr1, len(arr1), type(arr1))
list2 = arr1.tolist()
print(list2, len(list2), type(list2))
list2 = list(arr1)  # not to be done
print(list2, len(list2), type(list2))
```

```python
# examining arrays
arr1 = np.array([11, 33, 22, 44, 55, 66])
print(arr1, len(arr1), type(arr1))
print(arr1.dtype, arr1.size, arr1.ndim, arr1.shape)
```

```python
arr2 = np.array([[11, 33, 22, 44], [55, 66, 77, 88]])
print(arr2, len(arr2), type(arr2))
print(arr2.dtype, arr2.size, arr2.ndim, arr2.shape)
```

```python
var1 = (10)
print(var1, type(var1))
var1 = (10,)  # singleton representation of a tuple
print(var1, type(var1))
```

```python
arr1 = np.array([11, 33, 22, 44], dtype="int8")
print(arr1, len(arr1), type(arr1), arr1.dtype)
arr2 = np.array([[11, 33, 22, 44], [55, 66, 77, 88]], dtype="int8")
print(arr2, len(arr2), type(arr2), arr2.dtype)
```

```python
# create special arrays
print(np.zeros(10))
print(np.zeros(10, dtype="int8"), np.zeros(10).astype(int))
print(np.zeros(10) + 5) # vector & scalar operation -> vector
print(np.zeros((3, 6)))
print(np.zeros((3, 6)).astype(int))
```

```python
# create special arrays
print(np.ones(10))
print(np.ones(10, dtype="int8"), np.ones(10).astype(int))
print(np.ones(10) * 5) # vector & scalar operation -> vector
print(np.ones((3, 6)))
print(np.ones((3, 6)).astype(int))
```

```python
print(np.linspace(0, 1, 5)) # 0 to 1 (all inclusive) 5 equi-distant numbers
print(np.linspace(0, 10, 11))
print(np.linspace(0, 10, 11).astype(int))
```

```python
print(np.logspace(0, 3, 4))
print(np.logspace(0, 3, 4, base=10))
print(np.logspace(0, 3, 4, base=2))
print(np.logspace(0, 10))  # by default 50 numbers
```

```python
list1 = range(10)
print(list1, type(list1), list(list1), type(list(list1)))
arr1 = np.arange(10)
print(arr1, len(arr1), type(arr1))
arr2 = np.arange(10).astype(float)
print(arr2, arr2.shape)
arr2 = np.arange(10, dtype="float")
print(arr2, arr2.shape)
```

## **reshaping**

```python
matrix = np.arange(10, dtype=float)
print(matrix, matrix.shape)
matrix = np.arange(10, dtype=float).reshape(2, 5)
print(matrix, matrix.shape)
matrix = np.arange(10, dtype=float).reshape(-1, 5)
print(matrix, matrix.shape)
print()
matrix = np.arange(10).astype(float).reshape(2, -1)
print(matrix, matrix.shape)
matrix = np.arange(10).reshape(-1, 5).astype(float)
print(matrix, matrix.shape)
```

```python
matrix = np.arange(12).reshape(2, 2, -1).astype(float)
print(matrix)
```

```python
matrix =np.arange(10).reshape(2, 5)
print(matrix)
print(matrix.flatten())  # multi-dimestional array to one dimensional array conversion
print(matrix.T)  # transpose of a matrix
print(matrix.transpose())  # transpose of a matrix
```
