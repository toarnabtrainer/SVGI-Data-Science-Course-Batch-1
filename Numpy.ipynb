## **Create arrays**

```python
# create arrays from lists
# remember arrays will have data of same data types
data1 = [11, 22, 44, 33, 66, 55]
print(data1, type(data1), len(data1), id(data1))
arr1 = np.array(data1)
print(arr1, type(arr1), len(arr1), id(arr1))
```

```python
# function with variable number of arguments
def funct1(*arg):  # arg is defined forcefully as tuple type
    print(arg, len(arg), type(arg))
funct1("Amit", "Developer")
funct1("Amit", "Developer", "Pune")
funct1("Amit", "Developer", "Pune", "IBM")
```

```python
# function with variable number of arguments
def funct2(**kwarg):  # kwarg is defined forcefully as dictionary type
    print(kwarg, len(kwarg), type(kwarg))
funct2(name="Amit", desig="Developer")
funct2(name="Amit", desig="Developer", location="Pune")
funct2(name="Amit", desig="Developer", location="Pune", company="IBM")
```

```python
# function with variable number of arguments
def funct3(*arg, **kwarg):
    print(arg, len(arg), type(arg))
    print(kwarg, len(kwarg), type(kwarg), "\n")

funct3("Amit", "Developer", location="Pune", company="IBM", salary=80000)
funct3("Amit", "Developer")
funct3(location="Pune", company="IBM", salary=80000)
```

## **reshaping**

```python
matrix = np.arange(10, dtype=float)
print(matrix, matrix.shape)
matrix = np.arange(10, dtype=float).reshape(2, 5)
print(matrix, matrix.shape)
matrix = np.arange(10, dtype=float).reshape(-1, 5)
print(matrix, matrix.shape)
print()
matrix = np.arange(10).astype(float).reshape(2, -1)
print(matrix, matrix.shape)
matrix = np.arange(10).reshape(-1, 5).astype(float)
print(matrix, matrix.shape)
```

```python
matrix = np.arange(12).reshape(2, 2, -1).astype(float)
print(matrix)
```

```python
matrix =np.arange(10).reshape(2, 5)
print(matrix)
print(matrix.flatten())  # multi-dimestional array to one dimensional array conversion
print(matrix.T)  # transpose of a matrix
print(matrix.transpose())  # transpose of a matrix
```

```python
matrix = np.arange(10, dtype = float).reshape(2, 5)
print(matrix)
print("Stored in the C-style order (Row major):")
cmatrix = matrix.copy(order = 'C')
print(cmatrix)
for item in np.nditer(cmatrix):
    print(item, end = ", ")
```

```python
matrix = np.arange(10, dtype = float).reshape(2, 5)
print(matrix)
print("Stored in the F-style order (Column major):")
cmatrix = matrix.copy(order = 'F')
print(cmatrix)
for item in np.nditer(cmatrix):
    print(item, end = ", ")
```

## **append, insert, delete and sort operations**

```python
matrix = np.arange(10)
print(matrix, len(matrix))
matrix = np.append(matrix, [11, 22, 33])
print(matrix, len(matrix))
matrix  = np.append(matrix, [111, 222])
print(matrix, len(matrix))
```

```python
matrix = np.arange(10)
print(matrix, len(matrix))
matrix = np.insert(matrix, 3, [111, 333, 222])
print(matrix)
```

```python
# index ->         0   1   2   3   4   5   6   7
matrix = np.array([11, 33, 66, 55, 88, 77, 99, 22])
print(matrix, len(matrix))
matrix = np.delete(matrix, [6, 1, 2])
print(matrix, len(matrix))
```

```python
matrix = np.array([11, 33, 66, 55, 88, 77, 99, 22])
print(matrix)
matrix1 = np.sort(matrix)
print(matrix1)
matrix2 = np.sort(-matrix)
print(matrix2)
matrix3 = -np.sort(-matrix)
print(matrix3)
matrix4 = np.sort(matrix)[::-1]
print(matrix4)
```

```python
arr1 = np.arange(12).reshape(3, 4)
print(arr1)
arr2 = np.arange(12, 24).reshape(3, 4)
print(arr2)
arr3 = np.concatenate((arr1, arr2))
print(arr3)
print("Joining two matrices along axis-0...")
arr3 = np.concatenate((arr1, arr2))
print(arr3)
print("Joining two matrices along axis-1...")
arr3 = np.concatenate((arr1, arr2), axis = 1)
print(arr3)
print(arr3.shape, arr3.ndim)
```

```python
print("Stack two arrays along axis-0...")
arr4 = np.stack((arr1, arr2), axis = 0)
print(arr4)
print("Stack two arrays along axis-1...")
arr4 = np.stack((arr1, arr2), axis = 1)
print(arr4)
print(arr4.shape, arr4.ndim)
```

## **slicing and indexing**

```python
# L to R index   0   1   2   3   4   5   6   7
# R to L index  -8  -7  -6  -5  -4  -3  -2  -1
arr1 = np.array([22, 44, 33, 55, 77, 66, 88, 99])
print(arr1)
print(arr1[2], arr1[-6], arr1[7], arr1[-1])
print(arr1[:4], arr1[4:], arr1[::-1])
```

```python
arr2 = np.arange(20).reshape(4, 5)
print(arr2)
print(arr2[1:3, 1:4])
arr2[[0, -1], :] = 0
arr2[:, [0, -1]] = 0
print(arr2)
```

## **views and copies**

```python
arr1 = np.arange(10)
print(arr1)
arr_view1 = arr1[3:6]
print(arr_view1)
arr_view1[:] = 100
arr1[4] = 777
print(arr1)
print(arr_view1)
```

```python
arr1 = np.arange(10)
print(arr1)
arr_view1 = arr1[3:6].copy()
print(arr_view1)
arr_view1[:] = 100
arr1[4] = 777
print(arr1)
print(arr_view1)
```

## **using boolean arrays**

```python
arr1 = np.array([100, 300, -200, 450, 650, 800, -230])
print(arr1)
print(arr1 < 0)
print(arr1[arr1 < 0])
print(~(arr1 < 0))
print(arr1[~(arr1 < 0)])
```

```python
arr1 = np.array([100, 300, -200, 450, 650, 800, -230])
print(arr1)
print((arr1 >= 400) & (arr1 <= 700))
print(arr1[(arr1 >= 400) & (arr1 <= 700)])
print(arr1[(arr1 < 400) | (arr1 > 700)])
```

```python
arr1 = np.array(['bob', 'tom', 'pat', 'bob', 'tom'])
print(arr1)
# replace all 'bob' occurrences by 'tim'
print(arr1 == 'bob')
arr1[arr1 == 'bob'] = 'tim'
print(arr1)
print(np.unique(arr1))
```

## **broadcasting**

```python
a1 = np.array([[  0,  0,  0],
               [ 10, 10, 10],
               [ 20, 20, 20],
               [ 30, 30, 30]])
a2 = np.array([[0], [10], [20], [30]])
print(a1, type(a1), a1.shape)
print(a2, type(a2), a2.shape)
```

```python
b1 = np.array([[  0,  1,  2],
               [  0,  1,  2],
               [  0,  1,  2],
               [  0,  1,  2]])
b2 = np.array([0, 1, 2])
print(b1, type(b1), b1.shape)
print(b2, type(b2), b2.shape)
```

```python
print(a1 + b1)
print(a2 + b1)
print(a1 + b2)
print(a2 + b2)
```

```python
print(np.add(a1, b1))
print(np.add(a1, b2))
print(np.add(a2, b1))
print(np.add(a2, b2))
```
